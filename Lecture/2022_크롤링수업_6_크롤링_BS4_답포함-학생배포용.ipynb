```python
# 실습 2
import requests
url = requests.get("http://www.kpu.ac.kr/contents/main/cor/kcollege.html")

from bs4 import BeautifulSoup
#html = BeautifulSoup(url.content, "html.parser", from_encoding="utf-8")
html = BeautifulSoup(url.content, "html.parser")
eee = html.select(".unit")
for e in eee:
    # print(e)
    korean = e.select_one("div.label.ko").text
    english = e.select_one("div.label.en").text
    homepage = e.select_one(".homepage").get("href")
    print(korean, english, homepage, sep="\t")
    # print("==================================================")
```

```python
# 실습 3
import requests
sourceCode = requests.get("https://sports.news.naver.com/index.nhn", headers={'User-Agent': 'Mozilla/5.0'})

from bs4 import BeautifulSoup
html = BeautifulSoup(sourceCode.content, "html.parser", from_encoding="utf-8")
eee = html.select(".today_section div.text_area")

for e in eee:
    title = e.select_one(".title").text
    news = e.select_one(".information span").text
    sports = e.select_one(".information span:nth-child(2)").text
    print(news, sports, title, sep="\t")
```
