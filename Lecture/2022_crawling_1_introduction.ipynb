```python
#diy 4
#함수 정의 : 입력한 숫자가 소수인지 아닌지 판단
def isPrimeNumber(num):
    for k in range(2, num):
        if num % k == 0:
            return False
    return True
# print(isPrimeNumber(23))
#n부터 10000까지 for/while 문 돌려서 함수에 호출
    #만약 소수이면 리스트에 추가
    #리스트 길이를 판단, m개면 break
primeNumberList = []
# n = 5
# m = 10
def getMPrimeNumbers(n, m):
    # # for문 쓰는 방식
    # primeNumberList = []
    # for i in range(n, 100000):
    #     result = isPrimeNumber(i)
    #     if result == True: #소수이면
    #         #리스트에 추가
    #         primeNumberList.append(i)
    #         #리스트 길이가 m이면 break
    #         if len(primeNumberList) == m:
    #             break
    # return primeNumberList

    # while문 쓰는 방식
    primeNumberList = []
    i = n
    while len(primeNumberList) < m:
        result = isPrimeNumber(i)
        if result == True: #소수이면
            #리스트에 추가
            primeNumberList.append(i)
        # print(i, len(primeNumberList),primeNumberList)
        i += 1
    return primeNumberList    
```

```python
def printDict(dictionary):
    for key in dictionary:
        print(key, "\t", dictionary.get(key))
    print()
```

```python
dictionary = {}
dictionary["breakfast"] = "cereal"
printDict(dictionary)
print(dictionary.get("a chim"))
print(dictionary.get("breakfast"))
```
